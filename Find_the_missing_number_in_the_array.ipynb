```python
#Bruteforce - O(n^2)
nums=[9,6,4,2,3,5,7,0,1]
n=len(nums)
for i in range(0, n+1):
  if i not in nums:
    print(i)
```

```python
#moderate - O(3n)
nums=[9,6,4,2,3,5,7,0,1]
n=len(nums)
freq={}
for i in range(0, n+1):
  freq[i]=0
for num in nums:
  freq[num]=1
for k,u in freq.items():
  if u==0:
    print(k)
```

```python
#optimal - O(n)
nums = [9,6,4,2,3,5,7,0,1]
n = len(nums)

missing = n*(n+1)//2 - sum(nums)
print(missing)
```
