```python
def topKFrequent(words, k):

    freq = {}

    for word in words:
        if word in freq:
            freq[word] += 1
        else:
            freq[word] = 1

    sorted_words = sorted(
        freq.keys(),
        key=lambda word: (-freq[word], word)
    )

    return sorted_words[:k]


words = input("Enter words separated by spaces: ").split()

k = int(input("Enter k: "))

result = topKFrequent(words, k)

print("\nTop K Frequent Words:")
print(result)
```
