```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

list1=ListNode(1,ListNode(2,ListNode(4)))
list2=ListNode(1,ListNode(3,ListNode(4)))
def mergetwosortedarrays(list1, list2):
  if not list1:
    return list2
  if not list2:
    return list1

  if list1.val<list2.val:
    list1.next=mergetwosortedarrays(list1.next, list2)
    return list1
  else:
    list2.next=mergetwosortedarrays(list1, list2.next)
    return list2

result=mergetwosortedarrays(list1, list2)
curr = result
while curr:
    print(curr.val, end=" -> ")
    curr = curr.next
print("None")
```
