```python
#O(nlogn)
a=[1,1,1,2,6,7,9]
b=[1,3,2,4,9,8,2]
a_set=set(a)
b_set=set(b)
res=[]
for num in a_set:
  res.append(num)
for num in b_set:
  if num not in a_set:
    res.append(num)
print(sorted(res))
```
