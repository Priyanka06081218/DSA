```python
#Bruteforce - O(n^2)
order = [3,1,2,5,6,4,7]
friends=[1,3,4]
res=[]
for i in range(0, len(order)):
  if order[i] in friends:
    res.append(order[i])
print(res)
```

```python
#Optimal - O(n)
order = [3,1,2,5,6,4,7]
friends=[1,3,4]
friends_set=set(friends)
res=[]
for num in order:
  if num in friends_set:
    res.append(num)
print(res)
```
