```python
from typing import List, Optional

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
def build_min_height_bst(nums: List[int]):
    def helper(left, right):
        if left > right:
            return None

        mid = (left + right) // 2

        root = TreeNode(nums[mid])
        root.left = helper(left, mid - 1)
        root.right = helper(mid + 1, right)

        return root

    return helper(0, len(nums) - 1)
```

```python
def is_balanced(root: Optional[TreeNode]):
    def check_height(node):
        if node is None:
            return 0

        left_height = check_height(node.left)
        if left_height == -1:
            return -1

        right_height = check_height(node.right)
        if right_height == -1:
            return -1

        if abs(left_height - right_height) > 1:
            return -1

        return max(left_height, right_height) + 1

    return check_height(root) != -1
```

```python
def is_bst(root: Optional[TreeNode]):
    def validate(node, low, high):
        if node is None:
            return True

        if not (low < node.val < high):
            return False

        return (validate(node.left, low, node.val) and
                validate(node.right, node.val, high))

    return validate(root, float('-inf'), float('inf'))
```

```python
def inorder(root):
    if root:
        inorder(root.left)
        print(root.val, end=" ")
        inorder(root.right)
```

```python
#Driver code
if __name__ == "__main__":
    print("Q1: Build Minimum Height BST")
    nums = [1, 2, 3, 4, 5, 6, 7]

    bst_root = build_min_height_bst(nums)

    print("Inorder Traversal of BST:")
    inorder(bst_root)
    print("\n")
    print("Q2: Check Balanced Tree")

    # Balanced Tree
    balanced_root = TreeNode(1)
    balanced_root.left = TreeNode(2)
    balanced_root.right = TreeNode(3)
    balanced_root.left.left = TreeNode(4)
    balanced_root.left.right = TreeNode(5)

    print("Balanced Tree:", is_balanced(balanced_root))   # True

    # Unbalanced Tree
    unbalanced_root = TreeNode(1)
    unbalanced_root.left = TreeNode(2)
    unbalanced_root.left.left = TreeNode(3)
    unbalanced_root.left.left.left = TreeNode(4)

    print("Unbalanced Tree:", is_balanced(unbalanced_root))
    print()

    print("Q3: Check BST")

    # Valid BST
    valid_bst = TreeNode(4)
    valid_bst.left = TreeNode(2)
    valid_bst.right = TreeNode(6)
    valid_bst.left.left = TreeNode(1)
    valid_bst.left.right = TreeNode(3)

    print("Valid BST:", is_bst(valid_bst))

    # Invalid BST
    invalid_bst = TreeNode(5)
    invalid_bst.left = TreeNode(2)
    invalid_bst.right = TreeNode(7)
    invalid_bst.left.right = TreeNode(6)

    print("Invalid BST:", is_bst(invalid_bst))
```
