```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


common = ListNode(8, ListNode(4, ListNode(5)))

headA = ListNode(4, ListNode(1, common))
headB = ListNode(5, ListNode(6, ListNode(1, common)))


visited = set()

curr = headA
while curr:
    visited.add(curr)
    curr = curr.next

curr = headB
found = False

while curr:
    if curr in visited:
        print(curr.val)
        found = True
        break
    curr = curr.next

if not found:
    print(None)
```
