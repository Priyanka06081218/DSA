```python
class Solution:
    def rob(self, nums):

        memo = {}

        def dfs(index):

            if index < 0:
                return 0
            if index == 0:
                return nums[0]

            if index in memo:
                return memo[index]

            pick = nums[index] + dfs(index - 2)

            not_pick = dfs(index - 1)

            memo[index] = max(pick, not_pick)
            return memo[index]

        return dfs(len(nums) - 1)


nums = [2, 7, 9, 3, 5]

obj = Solution()
ans = obj.rob(nums)

print("Maximum money that can be robbed:", ans)
```
