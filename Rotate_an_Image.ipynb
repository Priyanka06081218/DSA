### Using a new matrix

```python
def rotate_matrix(matrix):

    n = len(matrix)

    result = []

    for col in range(n):

        row = []

        for r in range(n - 1, -1, -1):
            row.append(matrix[r][col])

        result.append(row)

    return result


n = int(input("Enter matrix size: "))

matrix = []

print("Enter matrix rows:")

for i in range(n):
    row = list(map(int, input().split()))
    matrix.append(row)

print("\nOriginal Matrix:")
for row in matrix:
    print(row)

rotated = rotate_matrix(matrix)

print("\nRotated Matrix:")
for row in rotated:
    print(row)
```

### In-Place Rotation

```python
def rotate_matrix(matrix):

    n = len(matrix)

    # Transpose
    for i in range(n):
        for j in range(i + 1, n):

            matrix[i][j], matrix[j][i] = matrix[j][i], matrix[i][j]

    # Reverse each row
    for row in matrix:
        row.reverse()


# Input
n = int(input("Enter matrix size: "))

matrix = []

print("Enter matrix rows:")

for i in range(n):
    row = list(map(int, input().split()))
    matrix.append(row)

print("\nOriginal Matrix:")
for row in matrix:
    print(row)

rotate_matrix(matrix)

print("\nRotated Matrix:")
for row in matrix:
    print(row)
```
