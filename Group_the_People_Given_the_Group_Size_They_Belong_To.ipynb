```python
#Optimal-O(n)
groupsizes=[3,3,3,3,3,1,3]
groups={}
res=[]
for i, size in enumerate(groupsizes):
  groups.setdefault(size, []).append(i)
for size, people in groups.items():
  for i in range(0, len(people), size):
    res.append(people[i:i+size])
print(res)
```
