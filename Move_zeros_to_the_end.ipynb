```python
#Bruteforce
nums=[2,3,4,0,1,2,1,1,0,8,0,4,0,7,0,0]
n=len(nums)
temp=[]
for i in range(0, n):
  if nums[i]!=0:
    temp.append(nums[i])
nz=len(temp)
for i in range(0, nz):
  nums[i]=temp[i]

for i in range(nz, n):
  nums[i]=0

print(nums)
```

```python
#Optimal
def move_zeros(nums):
    n = len(nums)
    if n == 1:
        return
    i = 0
    while i < n:
        if nums[i] == 0:
            break
        i += 1

    if i == n:
        return
    j = i + 1
    while j < n:
        if nums[j] != 0:
            nums[i], nums[j] = nums[j], nums[i]
            i += 1
        j += 1

nums = [2,3,4,0,1,2,1,1,0,8,0,4,0,7,0,0]
move_zeros(nums)
print(nums)
```
