```python
def max_repeating(sequence, word):

    count = 0

    while word * (count + 1) in sequence:
        count += 1

    return count


# Input
sequence = input("Enter sequence: ")
word = input("Enter word: ")

# Function Call
result = max_repeating(sequence, word)

# Output
print("\nMaximum Repeating Count =", result)
```
