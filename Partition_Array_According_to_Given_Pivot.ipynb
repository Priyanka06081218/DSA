```python
nums=[9,12,5,10,14,3,10]
pivot=10
less = []
equal = []
greater = []
for num in nums:
  if num < pivot:
    less.append(num)
  elif num == pivot:
    equal.append(num)
  else:
    greater.append(num)

print(less + equal + greater)
```
