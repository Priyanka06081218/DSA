```python
#Bruteforce
nums1=[4,9,6,5,4,9,5,9]
nums2=[4,6,8,9,9]
lst=[]
for i in range(0, len(nums1)):
  for j in range(0, len(nums2)):
    if nums1[i]==nums2[j]:
      if nums1[i] not in lst:
        lst.append(nums1[i])
print(lst)
```

```python
#Optimal
nums1=[4,9,6,5,4,9,5,9]
nums2=[4,6,8,9,9]
lst=[]
nums1=set(nums1)
nums2=set(nums2)
for i in nums1:
  if i in nums2:
    lst.append(i)
print(lst)
```
