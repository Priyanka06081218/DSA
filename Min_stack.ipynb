```python
class MinStack:

    def __init__(self):
        self.stack = []
        self.minStack = []

    def push(self, val):
        self.stack.append(val)

        if not self.minStack or val <= self.minStack[-1]:
            self.minStack.append(val)

    def pop(self):
        if not self.stack:
            print("Stack Empty")
            return

        if self.stack[-1] == self.minStack[-1]:
            self.minStack.pop()

        self.stack.pop()

    def top(self):
        if not self.stack:
            return "Stack Empty"
        return self.stack[-1]

    def getMin(self):
        if not self.minStack:
            return "Stack Empty"
        return self.minStack[-1]


obj = MinStack()

obj.push(5)
obj.push(3)
obj.push(7)
obj.push(2)

print("Top:", obj.top())
print("Min:", obj.getMin())

obj.pop()

print("After pop")
print("Top:", obj.top())
print("Min:", obj.getMin())
```
