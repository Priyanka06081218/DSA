```python
def findDuplicates(nums):
  map={}
  duplicates=[]
  count=0
  for i in range(len(nums)):
    map[nums[i]]=map.get(nums[i], 0)+1

    if map[nums[i]]>1:
      duplicates.append(nums[i])
  return duplicates

nums=[6,2,1,3,2,1,4,6,]
findDuplicates(nums)
```
