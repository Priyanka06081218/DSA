```python
#Time Complexity - O(n^2)
def containsduplicateII(nums, k):
  for i in range(0, len(nums)):
    for j in range(i+1, len(nums)):
      if nums[i]==nums[j] and abs(i-j)<=k:
        return True
  return False

nums=[1,0,1,1]
k=1
containsduplicateII(nums, k)
```

```python
#Time complexity - O(n)
def containsduplicateII(nums, k):
  seen={}
  for i in range(0, len(nums)):
    if nums[i] in seen and i - seen[nums[i]]<=k:
      return True
    seen[nums[i]]=i
  return False

nums=[1,2,3,4,2,1]
k=1
containsduplicateII(nums, k)
```
