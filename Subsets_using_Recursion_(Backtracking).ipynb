```python
#Optimal

class Solution:
    def subsets_with_target(self, nums, target):

        result = []
        subset = []

        def dfs(index, total):

            if total == target:
                result.append(subset.copy())
                return

            if total > target or index == len(nums):
                return

            subset.append(nums[index])
            dfs(index + 1, total + nums[index])

            subset.pop()

            dfs(index + 1, total)

        dfs(0, 0)
        return result


nums = [5, 9, 7, 2, 4, 1, 6, 3]
target = 9

obj = Solution()
ans = obj.subsets_with_target(nums, target)

print("Subsets with target sum:")
for s in ans:
    print(s)
```

```python
#Bruteforce

class Solution:
    def subsets(self, nums):

        result = []
        subset = []

        def dfs(index):

            if index >= len(nums):
                result.append(subset.copy())
                return

            subset.append(nums[index])
            dfs(index + 1)

            subset.pop()
            dfs(index + 1)

        dfs(0)
        return result


nums = [5, 9, 7]

obj = Solution()
ans = obj.subsets(nums)

print("All subsets are:")
for s in ans:
    print(s)
```
