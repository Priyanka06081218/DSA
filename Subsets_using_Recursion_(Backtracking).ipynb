```python
class Solution:
    def subsets(self, nums):

        result = []
        subset = []

        def dfs(index):

            if index >= len(nums):
                result.append(subset.copy())
                return

            subset.append(nums[index])
            dfs(index + 1)

            subset.pop()
            dfs(index + 1)

        dfs(0)
        return result


nums = [5, 9, 7]

obj = Solution()
ans = obj.subsets(nums)

print("All subsets are:")
for s in ans:
    print(s)
```
