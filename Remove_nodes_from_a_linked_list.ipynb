```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def removeNodes(self, head):
        if not head or not head.next:
            return head
        head.next=self.removeNodes(head.next)
        if head.next and head.val < head.next.val:
            return head.next
        return head

head=(ListNode(5,ListNode(2,ListNode(12,ListNode(3,ListNode(5,ListNode(8)))))))

sol = Solution()
result = sol.removeNodes(head)
curr=result
while curr:
    print(curr.val, end=" -> ")
    curr = curr.next
print("None")
```
