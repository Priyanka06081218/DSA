```python
# Input matrix
matrix = [
    [1, 1, 1],
    [1, 0, 1],
    [1, 1, 1]
]

rows = []
cols = []

# Find positions of zeros
for i in range(len(matrix)):
    for j in range(len(matrix[0])):
        if matrix[i][j] == 0:
            rows.append(i)
            cols.append(j)

# Make entire rows zero
for r in rows:
    for j in range(len(matrix[0])):
        matrix[r][j] = 0

# Make entire columns zero
for c in cols:
    for i in range(len(matrix)):
        matrix[i][c] = 0

# Print result
for row in matrix:
    print(row)
```
