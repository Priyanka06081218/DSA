```python
def longestConsecutive(nums):
  nums_set=set(nums)
  longest=0

  for n in nums_set:
    if n-1 not in nums_set:
      length=1
      while n+length in nums_set:
        length+=1
      longest=max(longest, length)
  return longest

nums = [100, 4, 200, 1, 3, 2]
longestConsecutive(nums)
```
