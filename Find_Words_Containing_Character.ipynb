```python
#Bruteforce-O(n)
words=["leet", "code"]
x="e"
res=[]
for ch in range(len(words)):
  if x in words[ch]:
    res.append(ch)
print(res)
```

```python
words=["abc", "bcd", "aaaa", "cbc", "bda"]
x="a"
res=[]
for ch in range(len(words)):
  if x in words[ch]:
    res.append(ch)
print(res)
```

```python
#Using enumerate
words=["abc", "bcd", "aaaa", "cbc", "bda"]
x="a"
res = []
for i, w in enumerate(words):
  if w.find(x) != -1:
    res.append(i)
print(res)
```
