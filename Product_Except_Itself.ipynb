```python
def product(nums):
  total_product = 1

  for num in nums:
      total_product *= num

  res = []

  for num in nums:
      res.append(total_product // num)
  return res


nums = [1,2,3,4]
print(product(nums))
```

```python
def product_except_self(nums):

    n = len(nums)

    result = [1] * n

    # Prefix Product
    prefix = 1

    for i in range(n):
        result[i] = prefix
        prefix *= nums[i]

    # Suffix Product
    suffix = 1

    for i in range(n - 1, -1, -1):
        result[i] *= suffix
        suffix *= nums[i]

    return result


nums = list(map(int, input("Enter numbers: ").split()))

print(product_except_self(nums))
```
